## Introduction and EDA

The question I would like to focus on is can you create a model that classifies individuals as smokers or non-smokers based on the other available data from previous years? This kind of information could be really useful to health insurance companies as they determine their premium prices for the coming year. 

Even if you ask a person directly whether they smoke or not, there will always be some who either didn't properly understand the question or flat out lie about their smoking status in perhaps an attempt to avoid higher premiums. Health insurance companies need to be able to predict a rough margin of error to ensure that the premiums they charge are capable of covering the potential unexpected expenses the coverage needs to pay out. It is common knowledge that smoking can cause a number of detrimental health effects and can result in more frequent and more costly medical bills.

Given that medical data is strongly protected by privacy laws, very little information is often available to companies to make informed decisions. I found  a small dataset of health insurance information from https://www.kaggle.com/datasets/rajgupta2019/medical-insurance-dataset?select=Train_Data.csv.  Although the dataset is small, it does have some interesting variables that might help us predict whether that particular person smokes or not.

## Model Building

#### Split the cleaned data into a training and test set.

```python
# Load package
from sklearn.model_selection import train_test_split

# Split into training data and test data for the modeling
train, test = train_test_split(insuranceTrimD, test_size=0.2, random_state = 10)
```

### Smoker classification model feature selection

For the smoker classification model I am attempting, the `smoker_yes` column is my target and the other columns are my features.

```python
# Select all columns that are not `smoker_yes`
features1 = [i for i in train.columns if i not in ['smoker_yes']]

# Defining the features from the train and test values
xTrain1 = train[features1]
xTest1 = test[features1]

# Defining the target from the train and test values
yTrain1 = train[['smoker_yes']]
yTest1 = test[['smoker_yes']]
```

#### Build a default KNN classifier

The first model I will attempt to use is the default KNN classifier.  This model classifies data points based on their proximity to each other.  Since my smokers and non-smokers are fairly clustered together in some of my preliminary graphs, this seems like a good jumping off point.

```python
# Import packages
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
# Create standardizer to bring all variables to the same scale
standardizer = MinMaxScaler()

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)

# Create a pipeline
pipe = Pipeline([("standardizer", standardizer), ("classifier", knn)])
```

```python
# Apply default KNN classifer
pipe.fit(xTrain1, yTrain1)

# Determine accuracy of the model
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Calculate predicted output values for the training data
yPredTrain1 = pipe.predict(xTrain1)

# Find accuracy of the default KNN on the training data
resultTrain = confusion_matrix(yTrain1, yPredTrain1)
print('Confusion Matrix:')
print(resultTrain)
resultTrain1 = classification_report(yTrain1, yPredTrain1)
print('Classification Report:')
print (resultTrain1)
resultTrain2 = accuracy_score(yTrain1, yPredTrain1)
print('Accuracy:',resultTrain2)
```

An accuracy score of 0.97 is really quite good.  The confusion matrix shows that there are 33-40 misidentified smoker statuses in each group, but overall more are correctly identified than not (based on the self-identified information). Let's see how well this model works with the test data.

```python
# Calculate predicted output values for the test data
yPredTest1 = pipe.predict(xTest1)

# Find accuracy of the default KNN on the test data
resultTest = confusion_matrix(yTest1, yPredTest1)
print('Confusion Matrix:')
print(resultTest)
resultTest1 = classification_report(yTest1, yPredTest1)
print('Classification Report:')
print (resultTest1)
resultTest2 = accuracy_score(yTest1, yPredTest1)
print('Accuracy:',resultTest2)
```

When applied to the test data, the accuracy of the default KNN model did decrease a little from 0.97, but is still really good at 0.95 so accurately predicting 95% of the categories correctly.  Looks like the erroenously sorted values decreased by about half while the correctly labeled values decreased by only about 22-24% meaning that the proportional error increased and explains why the overall accuracy decreased.

Let's see how well a model works when it doesn't have the `charges` value since this would not be available until the end of the current year.

### Smoker classification model without `charges` feature selection

```python
# Drop `charges` column
trainNoCharge = train.drop('charges', axis=1)
testNoCharge = test.drop('charges', axis=1)

# Select all columns that are not `smoker_yes`
features2 = [i for i in trainNoCharge.columns if i not in ['smoker_yes']]

# Defining the features from the train and test values
xTrain2 = trainNoCharge[features2]
xTest2 = testNoCharge[features2]

# Defining the target from the train and test values
yTrain2 = trainNoCharge[['smoker_yes']]
yTest2 = testNoCharge[['smoker_yes']]
```

#### Build a new default KNN classifier without the `charges` column

```python
# Apply default KNN classifer
pipe.fit(xTrain2, yTrain2)

# Determine accuracy of the model

# Calculate predicted output values for the training data
yPredTrain2 = pipe.predict(xTrain2)

# Find accuracy of the default KNN on the training data
resultTrain = confusion_matrix(yTrain2, yPredTrain2)
print('Confusion Matrix:')
print(resultTrain)
resultTrain1 = classification_report(yTrain2, yPredTrain2)
print('Classification Report:')
print (resultTrain1)
resultTrain2 = accuracy_score(yTrain2, yPredTrain2)
print('Accuracy:',resultTrain2)
```

The accuracy score on the training data drops from around 0.97 to 0.91 when the `charges` column is omitted, but that is still a really high accuracy score for classifying patrons as smokers or non-smokers based on self-identified information.  We can see that the misidentified data points are greater than the model that had the charges value in it, particularly for the smokers (91 misidentified vs 251 correctly identified).  Overall though, the model is still fairly accurate and might still give us some useful information.

```python
# Calculate predicted output values for the test data
yPredTest2 = pipe.predict(xTest2)

# Find accuracy of the default KNN on the test data
resultTest = confusion_matrix(yTest2, yPredTest2)
print('Confusion Matrix:')
print(resultTest)
resultTest1 = classification_report(yTest2, yPredTest2)
print('Classification Report:')
print (resultTest1)
resultTest2 = accuracy_score(yTest2, yPredTest2)
print('Accuracy:',resultTest2)
```

Looks like the accuracy of the non-charges model starts to really deterioate when applied to the test data.  Not only did the accuracy drop from 0.91 to 0.84, but also if we look at the confusion matrix, the number of misidentified smokers almost doubly outweighs the number that were correctly identified.  Based on this fact, I cannot in good conscious support this model despite its high accuracy value.

##### Summary

One flaw I have realized in the project idea is that I have no actual way of saying if a self-identified non-smoker is lying about their smoking status.  Thus the error rates I am seeing in my models could be due to model error, smokers lying, or a combination of both.  Unfortunately I don't have a good way to tell what the error's source is.  Going off of my initial idea, I could claim that all the error from the model is due to smokers lying, but that would be unfair to assume that the model is 100% accurate. The second model without the `charges` value, clearly shows how important this variable was in classifying the smoker categories. ### Predicting charges from supplied data

Let's do one more model just for grins and giggles.  This time let's see if we can predict the `charges` value  from the other data we have.  This could be beneficial for insurance companies to predict their expenditures based on the biographic data their patrons provide.

Since I am not trying to classify the data points into categories like before, but rather I am trying to predict their actual value, I am going to first try a linear regression. Let's explore the correlations between the variables first.

```python
# Create a correlation coefficient matrix for the whole training dataframe
matrix = train.corr()

# Create heatmap visualization of data matrix
sns.heatmap(matrix, annot=True)
```

Unsurprisingly, `smoker` status is the most highly correlated variable to charges with a 0.77 correlation value.  The next highest values are `age` at 0.3 and `bmi` at 0.21.  To avoid potentially overfitting the model with extra variables, I will only include these 3 highest correlated ones in my model. First let's start though with the single `smoker_yes` variable.

##### Single Linear Regression (Smoker vs Charges)

```python
# Load package
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Define linear regression model
x1 = train['smoker_yes'].values.reshape(-1,1)
x2 = test['smoker_yes'].values.reshape(-1,1)

y1 = train['charges'].values.reshape(-1,1)
y2 = test['charges'].values.reshape(-1,1)

model.fit(x1, y1)

predictionsTRAIN = model.predict(x1)
predictionsTEST = model.predict(x2)
```

```python
# Load package
from sklearn import metrics

# R2
TrainR2 = model.score(x1, y1)
TestR2 = model.score(x2, y2)

print("Train R2 = ", TrainR2, "Test R2 = ", TestR2)
```

Both the train and test data produces approximately the same R2 value as the original data (0.77 squared = 0.6) which would preliminarily suggest that the model faithfully replicates the trend seen in the data.  It also suggests that a linear model is just an ok fit for this data as it explains roughly 60% of the variance seen in the data.

```python
# Root Mean Squared Error (RMSE)
TrainRMSE = np.sqrt(metrics.mean_squared_error(y1, predictionsTRAIN))
TestRMSE = np.sqrt(metrics.mean_squared_error(y2, predictionsTEST))

print("Train RMSE = ", TrainRMSE, "Test RMSE = ", TestRMSE)
```

The RMSE of the test data is roughly the same as that of the training data, but is slightly lower indicating a slightly tighter fit.  Overall the RMSE value suggests an error of about $6800 off which is fairly significant when it comes to an individual patron's cost but perhaps in the grand scheme of the insurance business might not be a bad error rate.

```python
# Mean Absolute Error (MAE)
TrainMAE = metrics.mean_absolute_error(y1, predictionsTRAIN)
TestMAE = metrics.mean_absolute_error(y2, predictionsTEST)
print("Train MAE = ", TrainMAE, "Test MAE = ", TestMAE)
```

Once again the two values are fairly similar between the training and test data sets though the test set has a slightly better fit at 4980 dollars as opposed to 5021. Overall this linear regression model seems to be a ok fit for this data, but has a lot of room for improvement.

##### Multiple linear regression

```python
# Select variable columns for multiple linear regression
# Defining the features from the train and test values
xTrain3 = train[['smoker_yes', 'bmi', 'age']]
xTest3 = test[['smoker_yes', 'bmi', 'age']]

# Defining the target from the train and test values
yTrain3 = train[['charges']]
yTest3 = test[['charges']]
```

```python
# Define linear regression model
model.fit(xTrain3, yTrain3)

predictionsTRAIN = model.predict(xTrain3)
predictionsTEST = model.predict(xTest3)
```

```python
# R2
TrainR2 = model.score(xTrain3, yTrain3)
TestR2 = model.score(xTest3, yTest3)

print("Train R2 = ", TrainR2, "Test R2 = ", TestR2)
```

The multiple linear regression model with a R2 value hovering around 0.7 does better than the single regression (0.6).  Both the train and test data produce roughly the same value and explain roughly 70% of the variance seen in the data.  Not too bad.

```python
# Root Mean Squared Error (RMSE)
TrainRMSE = np.sqrt(metrics.mean_squared_error(yTrain3, predictionsTRAIN))
TestRMSE = np.sqrt(metrics.mean_squared_error(yTest3, predictionsTEST))

print("Train RMSE = ", TrainRMSE, "Test RMSE = ", TestRMSE)
```

The RMSE of the test model is roughly the same as that of the training data, but is slightly lower indicating a slightly tighter fit.  Overall the RMSE value of 5800 dollars is improved from the single regression value of 6800, but is still perhaps a slight bit too high.
