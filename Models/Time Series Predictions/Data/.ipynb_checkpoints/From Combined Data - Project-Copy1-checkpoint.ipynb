###  Creating Subsets for Analysis

- We will drop the 2022 year due to there only being a single value for that year.
- We will also create three different sub data sets:
    - Set One will run from 1961 - 2021 containing 'TempChangeC', 'Yield100GperHa', 'HectaresHarvested',' ExportValue1000USD', and 'ImportTons' since all of these variables are available that far back
        - This data set will help us address how changing temperatures impact almond crops.  It will also allow us to look at historical profit trends.
    - Set Two will run from 1991 - 2021 and add the 'ProdUSDperTonne' variable to Set One
        - This additional variable will help us look at the potential profits that an almond grower could take home from selling their crop 
    - Set Three will run from 2000 - 2020 and add the 'WaterStressPerct' variable as that variable starts at 2000
        - This last variable allows us to see if there are any significant impacts on harvest due to water scarcity.  This variable is unfortunately only available for such a short period of time.

```python
# Extract rows for 1961 - 2021
SetOne = CombinedData[CombinedData['Year'] <= 2021]
# Drop incomplete columns
SetOne = SetOne.drop(['ProdUSDperTonne','WaterStressPerct'], axis=1)

# Make sure the variables are numeric
SetOne['Yield100GperHa'] = pd.to_numeric(SetOne['Yield100GperHa'])
SetOne['HectaresHarvested'] = pd.to_numeric(SetOne['HectaresHarvested'])
SetOne['ExportValue1000USD'] = pd.to_numeric(SetOne['ExportValue1000USD'])
SetOne['ImportTons'] = pd.to_numeric(SetOne['ImportTons'])
SetOne
```

## Train and Test Datasets

```python
# Train-Test Split for SetOne (1961 - 2021, 60 years)
train_SetOne = SetOne[SetOne['Year'] < 2011] # 50 years
test_SetOne = SetOne[SetOne['Year'] >= 2011] # 10 years

# Train-Test Split for SetTwo (1991 - 2021, 30 years)
train_SetTwo = SetTwo.loc[SetTwo['Year'] < 2016] # 25 years 
test_SetTwo = SetTwo.loc[SetTwo['Year'] >= 2016] # 5 years
```

## Standarization

```python
from sklearn.preprocessing import StandardScaler # Define instance of standard scaler
scaler = StandardScaler()

# Transform the data using the standard scaler and convert the resulting output back into a dataframe
train_SetOne = pd.DataFrame(scaler.fit_transform(train_SetOne), columns = train_SetOne.columns)
test_SetOne = pd.DataFrame(scaler.fit_transform(test_SetOne), columns = test_SetOne.columns)
train_SetTwo = pd.DataFrame(scaler.fit_transform(train_SetTwo), columns = train_SetTwo.columns)
test_SetTwo = pd.DataFrame(scaler.fit_transform(test_SetTwo), columns = test_SetTwo.columns)
```

## Time Series Forecasting Models

```python
# Load Necessary libraries for models 
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm
import itertools
from statsmodels.tools.sm_exceptions import ConvergenceWarning

# Disable the specific warning
warnings.filterwarnings("ignore", message="Maximum Likelihood optimization failed to converge. Check mle_retvals", category=ConvergenceWarning)
warnings.filterwarnings("ignore", message="An unsupported index was provided and will be ignored when e.g. forecasting.", category=UserWarning)
```

```python
# Determine best ARIMA Order for SetOne
best_aic_SetOne = np.inf
best_order_SetOne = None

for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = sm.tsa.ARIMA(SetOne['Year'], order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic_SetOne:
                    best_aic_SetOne = results.aic
                    best_order_SetOne = (p, d, q)
            except:
                continue

print('Best ARIMA order for SetOne:', best_order_SetOne)
print('Best AIC for SetOne:', best_aic_SetOne)
```

```python
# Determine best ARIMA Order for SetTwo
best_aic_SetTwo = np.inf
best_order_SetTwo = None

for p in range(3):
    for d in range(2):
        for q in range(3):
            try:
                model = sm.tsa.ARIMA(SetTwo['Year'], order=(p, d, q))
                results = model.fit()
                if results.aic < best_aic_SetTwo:
                    best_aic_SetTwo = results.aic
                    best_order_SetTwo = (p, d, q)
            except:
                continue

print("Best ARIMA order for SetTwo:", best_order_SetTwo)
print("Best AIC for SetTwo:", best_aic_SetTwo)
```

```python
# ARIMA models for SetOne
model_SetOne = {}
results_SetOne = {}

for col in train_SetOne.columns:
    model_SetOne = sm.tsa.ARIMA(train_SetOne[col], order=(2, 1, 1))
    results_SetOne[col] = model_SetOne.fit()

# ARIMA models for SetTwo
model_SetTwo = {}
results_SetTwo = {}

for col in train_SetTwo.columns:
    model_SetTwo = sm.tsa.ARIMA(train_SetTwo[col], order=(2, 1, 2))
    results_SetTwo[col] = model_SetTwo.fit()

# Print the results for SetOne
for col, result in results_SetOne.items():
    print(f"Results for SetOne {col}: ")
    print(result.summary())
    print("---------------------------")

# Print the results for SetTwo
for col, result in results_SetTwo.items():
    print(f"Results for SetTwo{col}: ")
    print(result.summary())
    print("---------------------------")
```

```python
# Forecast and Evaluate
forecast_SetOne = {}
forecast_SetTwo = {}

# Forecast for SetOne
for col, model in results_SetOne.items():
    forecast_SetOne[col] = model.forecast(steps=11)

# Forecast for SetTwo
for col, model in results_SetTwo.items():
    forecast_SetTwo[col] = model.forecast(steps=6) 
```

## Compare Performance Metrics

```python
#drop 'Year' column
train_SetOne.drop(['Year'], axis=1, inplace=True)
test_SetOne.drop(['Year'], axis=1, inplace=True)

train_SetTwo.drop(['Year'], axis=1, inplace=True)
test_SetTwo.drop(['Year'], axis=1, inplace=True)
```

```python
# Compare Performance
mae_SetOne = {}
mse_SetOne = {}
rmse_SetOne = {}
mae_SetTwo = {}
mse_SetTwo = {}
rmse_SetTwo = {}

# Calculate metrics for SetOne
for col in test_SetOne.columns:
    mae_SetOne[col] = mean_absolute_error(test_SetOne[col], forecast_SetOne[col])
    mse_SetOne[col] = mean_squared_error(test_SetOne[col], forecast_SetOne[col])
    rmse_SetOne[col] = np.sqrt(mse_SetOne[col])
    
# Calculate metrics for SetTwo
for col in test_SetTwo.columns:
    mae_SetTwo[col] = mean_absolute_error(test_SetTwo[col], forecast_SetTwo[col])
    mse_SetTwo[col] = mean_squared_error(test_SetTwo[col], forecast_SetTwo[col])
    rmse_SetTwo[col] = np.sqrt(mse_SetTwo[col])    
    
# Print the performance metrics for SetOne
print("Performance metrics for SetOne:")
for col in test_SetOne.columns:
    print(f"Column '{col}':")
    print(f"MAE: {mae_SetOne[col]}")
    print(f"MSE: {mse_SetOne[col]}")
    print(f"RMSE: {rmse_SetOne[col]}\n")

# Print the performance metrics for SetTwo
print("Performance metrics for SetTwo:")
for col in test_SetTwo.columns:
    print(f"Column '{col}':")
    print(f"MAE: {mae_SetTwo[col]}")
    print(f"MSE: {mse_SetTwo[col]}")
    print(f"RMSE: {rmse_SetTwo[col]}\n")
```

## Results for MAE Model for SetOne and SetTwo

```python
# Convert to list MAE
mae_SetOne_values = list(mae_SetOne.values())
mae_SetTwo_values = list(mae_SetTwo.values())
```

```python
# Plot the individual MAE values for SetOne
plt.figure(figsize=(10, 6))
plt.bar(list(mae_SetOne.keys()), mae_SetOne_values, color='blue', label='SetOne')
plt.xlabel('Column Names')
plt.ylabel('MAE')
plt.title('MAE for SetOne')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the individual MAE values for  SetTwo
plt.figure(figsize=(10, 6))
plt.bar(list(mae_SetTwo.keys()), mae_SetTwo_values, color='orange', label='SetTwo')
plt.xlabel('Column Names')
plt.ylabel('MAE')
plt.title('MAE for SetTwo')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Compare mean MAE values for SetOne and SetTwo
mean_mae_SetOne = np.mean(mae_SetOne_values)
mean_mae_SetTwo = np.mean(mae_SetTwo_values)

# Plot the mean RMSE comparison for SetOne and SetTwo
N = 5
ind = np.arange(N)  # the x locations for the groups
width = .4 # the width of the bars
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.bar(list(mae_SetTwo.keys()), mae_SetTwo_values, width, color='orange', label='SetTwo')
ax.bar(ind+width, mae_SetOne_values, width, color='blue', label='SetOne')

# add some labels
plt.xlabel('Datasets')
plt.ylabel('Mean MAE')
plt.title('Mean MAE Comparison between SetOne and SetTwo')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Choose the better RMSE model
if mean_mae_SetOne < mean_mae_SetTwo:
    print("MAE Model for SetOne is better.")
else:
    print("MAE Model for SetTwo is better.")
```
